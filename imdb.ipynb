#### Figure out something "interesting" using the actors data!

```python
movies.sort_values('title', ascending=False).groupby(['actors_list', 'title']).star_rating.mean()
movies.groupby(['actors_list','title']).star_rating.mean()
```

```python
sns.boxplot(movies.star_rating, movies.actors_list)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
movies.groupby(['title', 'star_rating']).count()['actors_list'].unstack().plot(ax=ax)
ax.set_xlabel('Stars')
ax.set_ylabel('Actors')
```
